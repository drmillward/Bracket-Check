```python
def bracket_check(textstring):
    
    for line in textstring:
        
#       set an empty list to count the total number of open brackets
        openbrac = []
#       set an empty list to count each individual type of bracket
        brac1 = []
        brac2 = []
        brac3 = []
        brac4 = []
#       set a check which will be changed if a bracket is incorrect
        check = 0

        for sym in line:
            
#           if the symbol is an open bracket, append a number to the total count, and the same number to the specific bracket list
            if sym == '(':
                openbrac.append(len(openbrac)+1)
                brac1.append(openbrac[-1])
#           if the symbol is a closed bracket, and no corresonding open bracket has already been, switch check to one and break loop
            elif sym == ')' and len(brac1) == 0:
                check = 1
                break
#           if closed bracket and the last item of the total list does not match the last of the specfic list; 
#           i.e. a new bracket has opened and not closed again since the correspnding open bracket;
#           change check to 1 and break the loop
            elif sym == ')' and openbrac[-1] != brac1[-1]:
                check = 1
                break
#           if closed bracket and the last item of the total brackets matched the last of the specific list;
#           i.e. no new backets have opened since the corresponding open bracket and any that have have closed again;
#           remove the last item from both the total and specific lists
            elif sym == ')' and openbrac[-1] == brac1[-1]:
                openbrac.pop()
                brac1.pop()
                
#           repeat these checks for all other types of brackets:

            if sym == '[':
                openbrac.append(len(openbrac)+1)
                brac2.append(openbrac[-1])
            elif sym == ']' and len(brac2) == 0:
                check = 1
                break
            elif sym == ']' and openbrac[-1] != brac2[-1]:
                check = 1
                break
            elif sym == ']' and openbrac[-1] == brac2[-1]:
                openbrac.pop()
                brac2.pop()

            if sym == '{':
                openbrac.append(len(openbrac)+1)
                brac3.append(openbrac[-1])
            elif sym == '}' and len(brac3) == 0:
                check = 1
                break
            elif sym == '}' and openbrac[-1] != brac3[-1]:
                check = 1
                break
            elif sym == '}' and openbrac[-1] == brac3[-1]:
                openbrac.pop()
                brac3.pop()

            if sym == '<':
                openbrac.append(len(openbrac)+1)
                brac4.append(openbrac[-1])
            elif sym == '>' and len(brac4) == 0:
                check = 1
                break
            elif sym == '>' and openbrac[-1] != brac4[-1]:
                check = 1
                break
            elif sym == '>' and openbrac[-1] == brac4[-1]:
                openbrac.pop()
                brac4.pop()
        
#       if, when all checks are completed, all open brackets have closed, and the check has not been triggered;
#       print 1 to indicate the brackets on the line are correct
        if len(openbrac) == 0 and check == 0:
            print(1)
#       otherwise, print 0 to indicate there is some error in the brackets on the line
        else:
            print(0)
```

```python
rawdata = open('bracketcheck.txt', 'r')

bracket_check(rawdata)

rawdata.close()
```
